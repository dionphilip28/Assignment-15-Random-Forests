### Problem Statement: A cloth manufacturing company is interested to know about the segment or attributes causes high sale. 

```python
import pandas as pd
import matplotlib.pyplot as plt
#from sklearn import datasets  
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn import preprocessing
```

```python
company = pd.read_csv('Company_data.csv')
```

```python
company
```

```python
company['Sales'] = np.where(company['Sales']>=8,'High','Low')
```

```python
company
```

```python
label_encoder = preprocessing.LabelEncoder()
company['ShelveLoc']=label_encoder.fit_transform(company['ShelveLoc'])
company['Urban']=label_encoder.fit_transform(company['Urban'])
company['US']=label_encoder.fit_transform(company['US'])
company['Sales']=label_encoder.fit_transform(company['Sales'])
```

```python
company
```

```python
company['Sales'].value_counts()
```

```python
array = company.values
```

```python
array
```

```python
X = array[:,1:]
Y = array[:,0]
```

```python
# Random Forest Classification
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, oob_score=True, max_features=3)
model.fit(X, Y)
```

```python
# Accuracy 
accuracy = model.score(X, Y)
print(accuracy*100)
```

```python
model.estimators_
```

```python
model.oob_score_
```

```python
model.classes_
```

```python
company['Sales_pred']=model.predict(X)
company[['Sales','Sales_pred']]
```

```python
company['Sales_pred'].value_counts()
```

```python
pd.crosstab(company['Sales'],company['Sales_pred'])
```

```python
# Confusion Matrix for the model accuracy
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(company['Sales'],company['Sales_pred'])
confusion_matrix
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
# Bagging
kfold = KFold(n_splits=10, random_state=7, shuffle=True)
cart = DecisionTreeClassifier()
model1 = BaggingClassifier(base_estimator=cart, n_estimators=100, random_state=7)
results1 = cross_val_score(model1, X, Y, cv=kfold)
print(results1.mean()*100)
```

```python
# AdaBoost Classification
num_trees = 100
kfold = KFold(n_splits=10, random_state=7,shuffle=True)
model2 = AdaBoostClassifier(n_estimators=num_trees, random_state=7)
results2 = cross_val_score(model2, X, Y, cv=kfold)
print(results2.mean()*100)
```
